### Aggregation (집계)

```python
# aggregation 기능은 데이터 탐색 단계에서 기초 통계량을 추출할 때 사용된다.
# 각 요소의 합
mat.sum()
```

### Boolean array 활용법

```python
# 비교
mat > 10
```

```python
# 비교 (논리 연산자 복합적 사용)
(mat > 5) & (mat < 10)
```

```python
# Boolean Array를 이용한 element 선택 - 1
mat[mat > 10]
```

```python
# Boolean Array를 이용한 element 선택 - 2
mat[(mat > 5) & (mat < 10)]
```

```python
# Boolean Array를 이용한 element 선택 후 aggregation(집계) - 1
# 조건식이 True인 element의 갯수
np.sum(mat > 10)
```

```python
# Boolean Array를 이용한 element 선택 후 aggregation(집계) - 2
# select 후 sum()
np.sum(mat[mat > 10])
```

```python
#데이터 타입 확인
type(df)
```

```python
# 기본 정보 확인
df.info()
```

```python
# DataFrame은 Index라는 추가적인 정보를 가진다.
df.index
```

```python
# DataFrame은 Column이라는 추가적인 정보를 가진다.
df.columns
```

```python
# DataFrame 요소 선택
print(df.iloc[2, 2])
print(df.loc['샘플1', '꽃잎 길이'])
df['꽃잎 길이'], ['샘플1']
```

# Matplotlib 라이브러리

```python
#라이브러리 로드
import matplotlib.pyplot as plt

# plotting하기 위한 샘플데이터 생성
X = np.linspace(0, 100, 20) # 0~100까지 균일하게 20point
Y1 = X
Y2 = X**2
```

```python
# 그래프 생성
plt.plot(X, Y1)
```

```python
# stateless 하게 plotting
fig, axes = plt.subplots(figsize=(10, 5))
plt.title('Graph test')
plt.xlabel('X')
plt.ylabel('Y1')
axes.plot(X, Y1)
```

```python
# stateful 하게 plotting
fig, axes = plt.subplots(figsize=(10, 5))
plt.title('Graph test')
plt.xlabel('X label')
plt.ylabel('$Y_2$ label')
plt.plot(X, Y2)
plt.show()
```
